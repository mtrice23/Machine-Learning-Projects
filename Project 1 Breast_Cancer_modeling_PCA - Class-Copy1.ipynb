1. Define DataFrame df using cancer dataset


```python
df = pd.DataFrame(cancer.data)
df.columns = ['mean radius', 'mean texture', 'mean perimeter', 'mean area',
                   'mean smoothness', 'mean compactness', 'mean concavity',
                   'mean concave points', 'mean symmetry', 'mean fractal dimension',
                   'radius error', 'texture error',
                   'perimeter error', 'area error', 'smoothness error',
                   'compactness error', 'concavity error','concave points error',
                   'symmetry error', 'fractal dimension error', 'worst radius',
                   'worst texture', 'worst perimeter', 'worst area', 'worst smoothness',
                   'worst compactness', 'worst concavity', 'worst concave points',
                   'worst symmetry', 'worst fractal dimension']
```

### change benign to 0 and malignant to 1

```python
def correct_label(target):
    if target == 0:
        return 1
    else:
        return 0
```

```python
df['target']=df['target'].apply(correct_label)
```

# Model

7. Define X as dataset with target coumn and y as target column.

```python
X=df.drop('target',axis=1)
y=df['target']
```

8. Split X_train, X_test, y_train, y_test with test_size=0.5, random_state=42

```python
X_train, X_test, y_train, y_test = \
    train_test_split(X, y, test_size=0.5,
                     random_state=42)
```

9. Standardize X_train

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
lr= LogisticRegression(solver='lbfgs')
```

10. Use lr to train the tarining dataset.

```python
lr=lr.fit(X_train,y_train)
lr
```

11. Use lr to predict and print confusion_matrix.

```python
pred=lr.predict(X_test)
```

```python
print(metrics.confusion_matrix(y_test, pred))
print(metrics.classification_report(y_test, pred))
```

12. Use random_state=101 to print print confusion_matrix.

```python
X_train1, X_test1, y_train1, y_test1 = \
    train_test_split(X, y, test_size=0.5,
                     random_state=101)
sc = StandardScaler()
X_train1 = sc.fit_transform(X_train1)
X_test1 = sc.transform(X_test1)
```

```python
lr1=lr.fit(X_train1,y_train1)
pred1=lr.predict(X_test1)
```

```python
print(metrics.confusion_matrix(y_test1, pred1))
print(metrics.classification_report(y_test1, pred1))
```

## Two random state show different results on different cases, especially on FN, which can mean overfit.  Therefore, I will use PCA to reduce the dimensionality of a data set consisting of many variables correlated with each other.

13. Define pca with n_components=2

```python
pca = PCA(n_components=2)
```

14. Train pca

```python
pca
```

```python
sc = StandardScaler()
X_train_std = sc.fit_transform(X_train)
X_test_std = sc.transform(X_test)
X_train_pca = pca.fit_transform(X_train_std)
X_test_pca = pca.transform(X_test_std)
```

```python
X_train_pca.shape
```

15. Plot pca , cmap='plasma'

```python
plt.figure(figsize=(8,6))
plt.ylim(-8.5,11)
c = X_train_pca[:, 1]
plt.scatter(X_train_pca[:, 0], X_train_pca[:, 1],c=c,cmap='plasma')
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component', fontdict={'fontsize':9})
plt.show()
```

16. Print pca confusion_matrix

```python
X_train_pca, X_test_pca, y_train2, y_test2 = \
    train_test_split(X, y, test_size=0.5,
                     random_state=42)
X_train_pca = pca.fit_transform(X_train_std)
X_test_pca = pca.transform(X_test_std)
lr2=lr.fit(X_train_pca,y_train2)
pred2=lr.predict(X_test_pca)
print(metrics.confusion_matrix(y_test2, pred2))
print(metrics.classification_report(y_test2, pred2))
```
